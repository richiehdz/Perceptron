```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.datasets import load_iris
```

```python
iris=load_iris(as_frame=True)
```

```python
df=pd.DataFrame(data=iris.data.iloc[:,[2,3]], columns=iris.feature_names[2:4])
df['target']=iris.target
print(df)
sns.pairplot(df, hue='target')
```

```python
i=df.index[df['target']==2].tolist()
df=df.drop(i)
sns.pairplot(df, hue='target')
```

```python
variablex=iris.data
variablex=iris.data.iloc[:,[2,3]]
variabley=(iris.target)
```

```python
variablex=variablex.drop(i)
variabley=variabley.drop(i)

print(f'El tama;o de la variable independiente es: {variablex.shape}')

print(f'El tama;o de la variable dependiente es: {variabley.shape}')
```

```python
x_train, x_test, y_train, y_test=train_test_split(variablex, variabley, test_size=0.2)
model=Perceptron(random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
f1=f1_score(y_test, y_pred, average='macro')
print(f'El modelo obtuvo un indice f1 score de: {f1}')
percent=model.score(x_test, y_test)
print(f'El modelo obtuvo un {percent*100} de precision para clasificar')
```
